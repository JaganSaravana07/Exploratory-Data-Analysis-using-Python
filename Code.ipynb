###Function to handle outliers using IQR method and capping/flooring

```python
def handle_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Identify outliers
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]

    # Apply flooring and capping
    df[column] = np.where(df[column] < lower_bound, lower_bound, df[column])
    df[column] = np.where(df[column] > upper_bound, upper_bound, df[column])

    return df, outliers, lower_bound, upper_bound

# Apply the function to 'winner_runs'
matches_df, winner_runs_outliers, winner_runs_lower_bound, winner_runs_upper_bound = handle_outliers(matches_df, 'winner_runs')

# Apply the function to 'winner_wickets'
matches_df, winner_wickets_outliers, winner_wickets_lower_bound, winner_wickets_upper_bound = handle_outliers(matches_df, 'winner_wickets')

print("Outliers in winner_runs:\n", winner_runs_outliers)
print("Lower Bound:", winner_runs_lower_bound)
print("Upper Bound:", winner_runs_upper_bound)

print("Outliers in winner_wickets:\n", winner_wickets_outliers)
print("Lower Bound:", winner_wickets_lower_bound)
print("Upper Bound:", winner_wickets_upper_bound)
```

We can clearly see that most of team toss wins and decided to "Field" first.

```python

# Filter matches where the team won the toss and chose to field
toss_won_and_fielded = matches_df[matches_df['toss_decision'] == 'field']

# Filter matches where the team won the toss and chose to bat
toss_won_and_batted = matches_df[matches_df['toss_decision'] == 'bat']

# Calculate the total number of matches where toss was won and fielded
total_toss_won_and_fielded = len(toss_won_and_fielded)

# Calculate the total number of matches where toss was won and batted
total_toss_won_and_batted = len(toss_won_and_batted)

# Calculate the number of matches won by the team that won the toss and chose to field
wins_after_fielding = len(toss_won_and_fielded[toss_won_and_fielded['winner'] == toss_won_and_fielded['toss_won']])

# Calculate the number of matches won by the team that won the toss and chose to bat
wins_after_batting = len(toss_won_and_batted[toss_won_and_batted['winner'] == toss_won_and_batted['toss_won']])

# Calculate the winning percentages
win_percentage_fielding = (wins_after_fielding / total_toss_won_and_fielded) * 100
win_percentage_batting = (wins_after_batting / total_toss_won_and_batted) * 100

# Prepare data for the pie chart
labels = ['Won Toss & Fielded', 'Won Toss & Batted']
sizes = [win_percentage_fielding, win_percentage_batting]
colors = ['gold', 'lightskyblue']

# Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=90)
plt.title("Winning Percentage Based on Toss Decision in IPL 2023")
plt.show()
```

The winning percentage for the team winning the toss and batting first was 54.7%, and the team winning the toss and field was 45.3%. So, it clearly shows that winning the toss and bat first team has more chance to win a match compared to the team winning the toss and field.

###Top Bowlers by Economy Rate

```python
# Calculate total runs given and total balls bowled by each bowler
bowler_stats = score_df.groupby('bowler').agg({'outcome': 'count', 'batsman_score': 'sum'}).reset_index()

# Rename columns for clarity
bowler_stats.columns = ['bowler', 'balls_bowled', 'runs_given']

# Calculate economy rate (runs per over, where over = 6 balls)
bowler_stats['economy_rate'] = (bowler_stats['runs_given'] / bowler_stats['balls_bowled']) * 6

# Round the economy rate to two decimal places
bowler_stats['economy_rate'] = bowler_stats['economy_rate'].round(2)

# Filter out bowlers with an economy rate of zero
bowler_stats = bowler_stats[bowler_stats['economy_rate'] > 0]

# Sort by economy rate to get the top 10 bowlers with the best economy rates
top_economy_rate_bowlers = bowler_stats.sort_values(by='economy_rate').head(10)

# Plotting
labels = top_economy_rate_bowlers['bowler'].to_numpy()
ind = np.arange(len(labels))
width = 0.9

fig, ax = plt.subplots(figsize=(15, 8))
rects = ax.bar(ind, top_economy_rate_bowlers['economy_rate'].to_numpy(), width=width, color='lightgreen')

ax.set_xticks(ind)
ax.set_xticklabels(labels, rotation='vertical')
ax.set_ylabel("Economy Rate")
ax.set_title("Top Bowlers - Best Economy Rates in 2023 IPL")
ax.set_xlabel('Bowler Names')

# Function to add labels to the bars
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects)

plt.show()
```

###Surprisingly, the three players in the row with the lowest economy rate compared to the rest on the list are "Matthew Short", "Tilak Varma", and "Praveen Dubey". The thing is "Matthew Short" is a Australian all rounder, "Tilak Varma" is a part time bowler, and "Praveen Dubey" was an actual bowler who bowls leg spin.

```python
# Convert the 'over' column to an integer to properly filter on it
score_df['over'] = score_df['over'].astype(int)

# Filter for wickets (outcome 'w')
wickets = score_df[score_df['outcome'] == 'w']

# Group by 'inningno' and 'over' and count the number of wickets
wickets_by_inning_and_over = wickets.groupby(['inningno', 'over']).size().reset_index(name='wickets')

# Calculate the number of wickets fallen in the first half (overs 1-10) and second half (overs 11-20) for both innings
wickets_first_half = wickets_by_inning_and_over[wickets_by_inning_and_over['over'] <= 10].groupby('inningno')['wickets'].sum()
wickets_second_half = wickets_by_inning_and_over[wickets_by_inning_and_over['over'] > 10].groupby('inningno')['wickets'].sum()

# Combine the results into a DataFrame for easier plotting
wickets_summary = pd.DataFrame({
    'First Half': wickets_first_half,
    'Second Half': wickets_second_half
}).reset_index()

# Fill NaN values with 0 (in case there are innings with no wickets in either half)
wickets_summary = wickets_summary.fillna(0)

print(wickets_summary)
```

```python
# Visualization
plt.figure(figsize=(10, 6))

# Assuming wickets_summary is a DataFrame with the correct data
labels = ['1st Innings', '2nd Innings']
first_half_wickets = wickets_summary['First Half'].values
second_half_wickets = wickets_summary['Second Half'].values

# Verify the shapes
print(f"First Half Wickets: {first_half_wickets}")
print(f"Second Half Wickets: {second_half_wickets}")

# Check lengths
if len(first_half_wickets) != len(labels) or len(second_half_wickets) != len(labels):
    raise ValueError("Mismatch in the number of elements between labels and data arrays.")

x = range(len(labels))

plt.bar(x, first_half_wickets, width=0.4, label='First Half', align='center', color='blue')
plt.bar(x, second_half_wickets, width=0.4, label='Second Half', align='edge', color='green')

plt.xlabel('Innings')
plt.ylabel('Number of Wickets')
plt.title('Number of Wickets Fallen in First and Second Half of IPL 2023')
plt.xticks(ticks=x, labels=labels)
plt.legend()
plt.show()
```
